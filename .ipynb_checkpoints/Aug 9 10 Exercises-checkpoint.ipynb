```python
def pt(start,end):
    print(f"move {start} --> {end}")
```

```python
#Tower of Hanoi
def hanoi(n,start,end):
    if (n==1):
        pt(start,end)
    else:
        other=6-(start+end)
        hanoi(n-1,start,other)
        pt(start,end)
        hanoi(n-1,other,end)
    return None
    
```

```python
hanoi(1,1,2)
```

```python
hanoi(2,1,3)
```

```python
hanoi(2,3,1)
```

```python
hanoi(10,3,1)
```

```python
def sum_num(L1):
    sum=0
    for i in L1:
        sum=sum+i
    return sum
```

```python
sum_num([1,2,3,4])
```

```python
def count_num(L1):
    count=0
    for i in L1:
        count=count+1
    return count
```

```python
count_num([1,2,3,4])
```

```python
#Compute mean
def mean_num(L1):
    mean=sum_num(L1)/count_num(L1)
    return mean
```

```python
mean_num([1,2,3])
```

```python
def standard_dev(L1):
    mean = mean_num(L1)
    sum_SD=0
    for i in L1:
        sum_SD = sum_SD+((i-mean)**2)
    count=count_num(L1)
    SD=(sum_SD/count)**0.5  
    return SD
```

```python
standard_dev([1,2,3,4,5])
```

```python
# find the x and y of the equations
def find_x_y(L1,L2):
    a1,b1,c1=L1
    a2,b2,c2=L2
    x=((c2*b1)-(b2*c1))/((b1*a2)-(b2*a1))
    y=(c1-(a1*x))/b1
    return x,y
    
```

```python
find_x_y([1,1,10],[2,-1,-15])
```

```python
#using *arg
def find_x_y(*args):
    if (len(args)==1):
        if(len(args[0])==2):
            a1,c1=args[0]
            x=c1/a1
            return x
        else:
            print("Please input in the form of [a,b]")
    elif (len(args)==2):
            a1,b1,c1=args[0]
            a2,b2,c2=args[1]
            if ((b1*a2)-(b2*a1)==0):
                raise ValueError("The system has no unique solution.")
            else:
                x=((c2*b1)-(b2*c1))/((b1*a2)-(b2*a1))
                y=(c1-(a1*x))/b1
                return x,y   
        
```

```python
## Compute root mean square error in n-dimensions

def check_kwarg_arg(*args, **kwarg):
    for i in args:
        print(f"{i}")
    for a,b in kwarg.items():
        print(f"{a}={b}")
```

```python
check_kwarg_arg(['pizza','burger'],['italian','chinese'],size='small',quantity='2')
```
