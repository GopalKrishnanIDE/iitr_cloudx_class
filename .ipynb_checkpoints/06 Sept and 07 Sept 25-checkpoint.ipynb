```python
#Name Combination exercise
def permute(chars):
    if len(chars)==1:
        return [chars]
    perms=permute(chars[:-1])
    last_char=chars[-1]
    new_perm=[]
    for perm in perms:
        for i in range(len(perm)+1):
            new_p=perm[i:]+last_char+perm[:i]
            new_perm.append(new_p)
    return new_perm
```

```python
def find_first_blank_position(sudoku):
    for row in range(9):
        for col in range(9):
            if sudoku[row][col] == 0:
                return row, col
    return None

def find_numbers_to_fill(sudoku, i, j):
    row_vals = set(sudoku[i])
    col_vals = set(sudoku[x][j] for x in range(9))
    box_row, box_col = 3 * (i // 3), 3 * (j // 3)
    box_vals = set(
        sudoku[r][c]
        for r in range(box_row, box_row + 3)
        for c in range(box_col, box_col + 3)
    )
    return set(range(1, 10)) - row_vals - col_vals - box_vals

def solve(sudoku):
    pos = find_first_blank_position(sudoku)
    if pos is None:
        print(sudoku)
        return True
    i, j = pos
    candidates = find_numbers_to_fill(sudoku, i, j)
    for candidate in candidates:
        sudoku[i][j] = candidate
        if solve(sudoku):
            return True
        sudoku[i][j] = 0
    return False
```

```python
#8 Queen

def is_safe(board, row, col):
#     print("is_safe", row, col)
    for i in range(8):
        if board[i][col] == True:
            return False
        
    #check diagonally and anti-diagonally
    # diagonal
    i, j = row, col
    while i >= 0 and j >= 0:
        if board[i][j] == True:
            return False
        i -= 1
        j -= 1

    # anti diagonal
    i, j = row, col
    while i >= 0 and j < 8:
        if board[i][j] == True:
            return False
        i -= 1
        j += 1
        
    return True
```

```python
def create_board():
    board = []
    for i in range(8):
        row = []
        for j in range(8):
            row.append(False)
        board.append(row)
    return board
```

```python
  
def find_place_to_put_next(board, row):
    result = []
    for j in range(8):
        if board[row][j] == False and is_safe(board, row, j):
            result.append(j)
    return result
    
# eight queens problem
def solve(board, num_of_queens=0):
    if num_of_queens == 8:
        print(board)
        return True
    positions = find_place_to_put_next(board, row=num_of_queens)
    print(positions)
    i = num_of_queens
    for j in positions:
        board[i][j] = True
        solve(board, num_of_queens=num_of_queens+1)
        board[i][j] = False
```

```python
class Animal:
    name = 'No name'
    hunger = 10
    def eat(self, food): #object level
        self.hunger -= food
        print(f"{self.name} ate {food}. Now, hunger: {self.hunger}")
    def speak(self):
        print(None)
```

```python
class Dog(Animal):
    def __init__(self, name, breed, age):
        self.name = name.upper()
        self.breed = breed
        self.age = age + 3
    def bark(self):
        print(f"{self.name} is barking")
    def speak(self):
        self.bark()
    def __repr__(s):
        return f"{s.name} / {s.age} / {s.breed}"
```

```python
class Node:
    
    def __init__(self, val):
        self.data=val
        self.L=None
        self.R=None
        
    def printtree(self):
        if self.L !=None:
            self.L.printtree()
        print(self.data)
        if self.R !=None:
            self.R.printtree()
    
    def minval(self):
        current_min = self.data

    # Check left subtree
        if self.L is not None:
            left_min = self.L.minval()
            if left_min < current_min:
                current_min = left_min

    # Check right subtree
        if self.R is not None:
            right_min = self.R.minval()
            if right_min < current_min:
                current_min = right_min

        return current_min
    
    def maxval(self):
        current_max = self.data

    # Check left subtree
        if self.L is not None:
            left_max = self.L.maxval()
            if left_max > current_max:
                current_max = left_max

    # Check right subtree
        if self.R is not None:
            right_max = self.R.minval()
            if right_max > current_max:
                current_max = right_max

        return current_max
    
```

```python
root = Node(10)
_1 = Node(-1)
_1.L = Node(15)
_1.R = Node(20)

_2 = Node(-30)
root.L = _1
root.R = _2
```
