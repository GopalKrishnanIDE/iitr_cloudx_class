```python
def slope_of_line(x1,y1,x2,y2):
    m=(y2-y1)/(x2-x1)
    return m
```

```python
def intercept_line(x1,y1,x2,y2):
    m =slope_of_line(x1,y1,x2,y2)
    intercept = y1-m*(x1)
    return intercept
```

```python
def calculated_y (slope,intercept,x):
    return (slope*x)+intercept
```

```python
## is point on line segment in 2D
def is_point_on_line_segment(x1,y1,x2,y2,x3,y3):
    slope = slope_of_line(x1,y1,x2,y2)
    intercept = intercept_line(x1,y1,x2,y2)
    cal_y = calculated_y (slope,intercept,x3)
    if (cal_y==y3):
        return True 
    else:
        return False
    
    
```

```python
is_point_on_line_segment(0,0,4,-4,2,-2)
```
