```python
class TwoDPointLine:
    def __init__ (self):
        print("Model is created, here we go")
        
    def fit (self,X,Y):
        self.x1,self.x2 = X[0][0],X[1][0]
        self.y1,self.y2 = Y
        self.m = (self.y1-self.y2)/(self.x1-self.x2)
        self.c = self.y1-self.x1*self.m
        return self.m,self.c
    def predict(self,Z):
        result =[]
        
        for i in range(len(Z)):
            result.append(self.m*Z[i][0]+self.c)
        return result
        
        
```

```python
model=TwoDPointLine()
```

```python
X = [[1],[2]]
Y=[5,6]
model.fit(X,Y)
```

```python
model.predict([[3],[4],[5]])
```

```python
class ThreePointPlane:
    def __init__(self):
        print("Model is created, here we go")

    def fit(self, X, y):
        # Build augmented matrix: [x1, x2, 1 | y]
        equations = [
            [X[i][0], X[i][1], 1, y[i]]
            for i in range(3)
        ]

        # Gaussian elimination with pivoting
        for i in range(3):
            # Find pivot row
            if equations[i][i] == 0:
                for j in range(i + 1, 3):
                    if equations[j][i] != 0:
                        equations[i], equations[j] = equations[j], equations[i]
                        break
                else:
                    raise ValueError("Cannot solve: zero pivot encountered.")

            # Normalize pivot row
            pivot = equations[i][i]
            equations[i] = [x / pivot for x in equations[i]]

            # Eliminate below
            for j in range(i + 1, 3):
                factor = equations[j][i]
                equations[j] = [
                    equations[j][k] - factor * equations[i][k]
                    for k in range(4)
                ]

        # Back-substitution
        c = equations[2][3]
        m2 = equations[1][3] - equations[1][2] * c
        m1 = equations[0][3] - equations[0][1] * m2 - equations[0][2] * c

        self.m1, self.m2, self.c = m1, m2, c
        print(f"Fitted: m1 = {m1:.4f}, m2 = {m2:.4f}, c = {c:.4f}")
        return m1, m2, c

    def predict(self, Z):
        return [self.m1 * x[0] + self.m2 * x[1] + self.c for x in Z]


```

```python
X = [[1, 2], [2, 4], [6, 7]]
y = [5, 6, 10]

model = ThreePointPlane()
model.fit(X, y)
print(model.predict([[3, 5]]))  # Output: [8.4286]
```

```python
#Square root recursion
def my_sqrt(n):
    print(n)
    sqrt_internal(n,0,n)
    
def sqrt_internal(n,start,end):
    mid=(start+end)/2
    print(f"mid : {mid}")
    if abs((mid*mid)-n)<=0.001:
        return mid
    if abs((mid*mid)>n):
        end=mid
        return sqrt_internal(n,start,end)
    else:
        start=mid
        return sqrt_internal(n,start,end)
```

```python
def pploops(s):
    tasks=[([],s)]
    while len(tasks)>0:
        letters,s1=tasks.pop()
        print(f"letters,s1: {letters,s1}")
        if(len(s1)==0):
            print(letters)
        for letter1 in s1:
            s2=s1.replace(letter1,"")
            letters1=letters.copy()
            letters1.append(letter1)
            new_task=letters1,s2
            tasks.append(new_task)
```

```python
def binary_search(a,b):#this will work only in sorted.
    #print(f"a:{a}, b:{b}")
    #a_=sorted(a) # it will consume lot of computational space. as it goes through the entire list.
    a_=a
    #print(f"a_ is {a_}")
    start=0
    end=len(a)-1
    
    if(end==-1):
     #   print("List is empty")
        return False
    
    while True:
        #print("-----------")
        mid=((start+end)//2)
        #print(f"start: {start},end: {end},mid: {mid}")
        #print(f"-->a_[mid]: {a_[mid]}")
        old_mid=mid
        #print(f"old_mid: {old_mid}")
        if(a_[mid]==b):
            
         #   print(f"found the match")
            return True
        elif (a_[mid]>b):
            
            end=mid-1
            
          #  print(f"a[mid]:{a_[mid]} is greater than b:{b}")
            
        elif (a_[mid]<b):
            start=mid+1
           # print(f"a[mid]:{a_[mid]} is less than b:{b}")
           
        if(old_mid==(start+end)//2):
            #print(f"found nothing")
            return False
        
        
        
```

```python
print(binary_search([], 5))  # Expected: False (list is empty)
```

```python
class NN:
    def __init__(self):
        print("Nearest Neighbour function")
    def fit(self, X,y):
        Z=[]
        for i in range(len(y)):
            Z.append(X[i]+[y[i]])
        self.Z=Z
        return 
    def predict(self,A):
        dist=[]
        for i in range(len(self.Z)):
            dist.append((A[0]-self.Z[i][0])**2+(A[1]-self.Z[i][1])**2)
        print(dist,"-----")
        min_dist=float('inf')
        for i in range(len(dist)):
            if(dist[i]<min_dist):
                min_dist=dist[i]
                min_index=i
        print(min_dist,min_index)    
        
        return self.Z[min_index][2]
```

```python
class KNN:
    def __init__(self, K):
        print("Nearest Neighbour function")
        self.K = K

    def fit(self, X, y):
        self.Z = [X[i] + [y[i]] for i in range(len(y))]

    def predict(self, A):
        # Compute distances to all training points
        dist = []
        for i in range(len(self.Z)):
            d = (A[0] - self.Z[i][0])**2 + (A[1] - self.Z[i][1])**2
            dist.append((d, self.Z[i][-1]))  # (distance, label)

        # Sort by distance and pick top K
        dist.sort(key=lambda x: x[0])
        k_nearest = dist[:self.K]

        print("K nearest distances and labels:", k_nearest)
        
        sum_k=0

        for i in range(len(k_nearest)):
            sum_k=sum_k+k_nearest[i][1]
        prediction=sum_k/len(k_nearest)
       
        return prediction
```

```python
X=[[1,2],[2,4],[6,7],[0,2]]
y=[5,6,7,8]
model=KNN(1)
```

```python
model.fit(X,y)
```

```python
model.predict([1.1,2])
```
