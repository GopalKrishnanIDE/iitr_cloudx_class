# Function returning function, used for LLM

```python
def fit (x1,y1,x2,y2):
    
    # Training phase
    
    m = (y2-y1)/(x2-x1)
    c=y1-m*x1
    print(f"Learnt: m {m} and c: {c}")
   
    # defining a model called line using the above training phase data
    # so it will return the new defined function as a result
    def line(x):
        print("line calling")
        return m*x + c
    return line
```

```python
model=fit (1,2,3,4)
```

```python
model(3)
```

```python
fit(2,3,4,5)
```

```python
model=fit(2,4,6,8)
```

```python
model(1)
```

```python
def power_factory(n):
    def power(x):
        return x ** n
    return power
```

```python
square=power_factory(2)
```

```python
square(2)
```

```python
cube=power_factory(3)
```

```python
square(3)
```

```python
cube(3)
```

```python
def cointoss():
    if random.random() <0.5:
        return 0
    else:
        return 1
```

```python
def measurement(f):
    lc =100000
    h=0
    for i in range (lc):
        ret=f()
        h+=ret
    ratio=h/lc
    print(f"H: {ratio}")
```

```python
measurement(cointoss)
```
