```python
def flatten_nested_list(A):
    c=[]
    print(A)
    for e in A:
        if(type(e)==list):
            c.extend(flatten_nested_list(e))
        else: 
            c.append(e)
        print(c)
    return c
```

```python
flatten_nested_list([1,[2,[3,6],3],[4],5])
```

```python
def sum_of_the_tree(A):
    sum=0
    for e in A:
        if(type(e)==list):
            sum=sum+sum_of_the_tree(e)
        else:
            sum=sum+e
    return sum
            
    
```

```python
#lsp
def mylsp(A):
    if type (A)!=list:
        return A 
    op= A[0]
    v1=mylsp(A[1])
    v2=mylsp(A[2])
    if op=='+':
        return v1+v2
    if op=='-':
        return v1-v2
    if op=='/':
        return v1/v2
    if op =='*':
        return v1*v2
    if op =='^':
        return v1**v2
    return 0
```

```python
def is_palindrome(arr,start,end):
    if start>=end:
        return True
    if arr[start]==arr[end]:
        return is_palindrome(arr,start+1,end-1)
    
    return False
```
